## Part 5
**Outlier Detection and Removal**

```python
# Question 1, Print out the Outliers in each column, use the IQR method of outlier detection.
import statistics as stats

def outliers_IQR (column):
    Q1 = column.quantile(.25)
    Q3 = column.quantile(.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    outliers = (column < lower_bound) | (column > upper_bound)
    return outliers 

# use outliers_IQR function for each column
pima_fixed_columns_df['PregOutlier'] = outliers_IQR(pima_fixed_columns_df['Pregnancies'])
pima_fixed_columns_df['GlucoseOutlier'] = outliers_IQR(pima_fixed_columns_df['Glucose'])
pima_fixed_columns_df['BloodPressureOutlier'] = outliers_IQR(pima_fixed_columns_df['BloodPressure'])
pima_fixed_columns_df['SkinThicknessOutlier'] = outliers_IQR(pima_fixed_columns_df['SkinThickness'])
pima_fixed_columns_df['InsulinOutlier'] = outliers_IQR(pima_fixed_columns_df['Insulin'])
pima_fixed_columns_df['BMIOutlier'] = outliers_IQR(pima_fixed_columns_df['BMI'])
pima_fixed_columns_df['DiabetesPedigreeFunctionOutlier'] = outliers_IQR(pima_fixed_columns_df['DiabetesPedigreeFunction'])
pima_fixed_columns_df['AgeOutlier'] = outliers_IQR(pima_fixed_columns_df['Age'])
# doesn't work for oultiers becasue of datatype boolean 

pima_fixed_columns_df.head()
```

```python
# Question 2, Use loc to remove outliers in each of the columns that have outliers, save this as pima_outlier_removed_df.
pima_outlier_removed_df = pima_fixed_columns_df.loc[~pima_fixed_columns_df['PregOutlier']]
pima_outlier_removed_df = pima_outlier_removed_df.loc[~pima_outlier_removed_df['GlucoseOutlier']]
pima_outlier_removed_df = pima_outlier_removed_df.loc[~pima_outlier_removed_df['BloodPressureOutlier']]
pima_outlier_removed_df = pima_outlier_removed_df.loc[~pima_outlier_removed_df['SkinThicknessOutlier']]
pima_outlier_removed_df = pima_outlier_removed_df.loc[~pima_outlier_removed_df['InsulinOutlier']]
pima_outlier_removed_df = pima_outlier_removed_df.loc[~pima_outlier_removed_df['BMIOutlier']]
pima_outlier_removed_df = pima_outlier_removed_df.loc[~pima_outlier_removed_df['DiabetesPedigreeFunctionOutlier']]
pima_outlier_removed_df = pima_outlier_removed_df.loc[~pima_outlier_removed_df['AgeOutlier']]
# does not work for outcome b/c bool datatype whoch we already removed the error from 

pima_outlier_removed_df.head()
```
